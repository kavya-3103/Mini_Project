```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize

target=[]
images =[]
flat_data =[]

DATADIR ='/content/drive/MyDrive/newimages'
CATEGORIES =['apple','banana']

for category in CATEGORIES:
  class_num =CATEGORIES.index(category)  #label encoding.
  path =os.path.join(DATADIR,category)  #create path to use all images 
  for img in os.listdir(path):
    img_array =imread(os.path.join(path,img))
    #print(img_array.shape)
    #plt.imshow(img_array)
    img_resized = resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data =np.array(flat_data)
target = np.array(target)
images =np.array(images)    
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(flat_data,target,test_size=0.2,random_state=123)
```

SVM


```python
# Normalization : Bringing value in a scale of 0-1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid ={'C':[0.1,1,10,100],
              'kernel':['linear','rbf']}


grid = GridSearchCV(SVC(),param_grid)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.svm import SVC
from sklearn import svm
model =svm.SVC(kernel='rbf',C=1,gamma='scale')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
